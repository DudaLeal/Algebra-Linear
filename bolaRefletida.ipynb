```
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from matplotlib import animation
from IPython.display import HTML


```

```
# Geração de pontos para formar uma reta
npoints = 200
x0 = np.zeros(npoints)
x1 = np.zeros(npoints)
v = np.array([2.,-7.])

p =np.array([0.,0.])
deltaT = 0.01
for k in range(1,npoints):
  x0[k] = deltaT*v[0] + x0[k-1]
  x1[k] = deltaT*v[1] + x1[k-1]

  # alpha = alpha + 0.1
```

```
plt.figure(figsize=(6,6))
plt.plot(x0,x1,'r.')
plt.xlim(-10,10)
plt.ylim(-10,10)
```

Animação abaixo!!

```
# primeiro, vamos apagar a antiga figure
plt.close()

# mesmas atribuicoes...
fig = plt.figure()
ax = plt.axes(xlim=(-10, 10), ylim=(-10, 10))
line, = ax.plot([], [], 'ro', lw=2)
npoints = 20
x0 = np.zeros(npoints)
x1 = np.zeros(npoints)
v = np.array([2.,-7.])
p = np.array([0.,0.])
alpha = 0
deltaT = 0.2
```

```
# funcao init, inicia somente com background
def init():
    line.set_data([],[])
    return line,

# funcao animate, chamada sequencialmente. Funciona da mesma forma que um for i in range(frames)
def animate(i):
    global x0, x1, alpha, v, p
    x0[i] = deltaT*v[0] + x0[i-1]
    x1[i] = deltaT*v[1] + x1[i-1]

    # Verifica se a bola atingiu o eixo x
    if x0[i] < -10 or x0[i] > 10:
        v[0] = -v[0]

    # Verifica se a bola atingiu o eixo y
    if x1[i] < -10 or x1[i] > 10:
        v[1] = -v[1]

    line.set_data(x0, x1)
    return line,

# chama o animator.  blit=True so redesenha as partes que de fato mudaram.
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=npoints, interval=40, blit=True)




# Creditos e agradecimentos a Jake Vanderplas
# email: vanderplas@astro.washington.edu
# website: http://jakevdp.github.com
```
